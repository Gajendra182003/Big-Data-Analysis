```python
Create a class Circle which has a method setRadius and another method printArea. The user will use setRadius to set the radius value and printArea to print the area of a circle.
```

```python
class Circle:
    def set_radius(self,radius):
        self.radius=radius

    def print_area(self):
        print(f"Area is {3.14*self.radius**2}")
```

```python
object=Circle()
r=float(input("Enter radius:"))
object.set_radius(r)
object.print_area()
```

```python
Create a class calculator which has four methods:
- add
- multiply
- subtract
- divide
which is called with two data members and will print the appropriate result.
#sample:
object.add(10,20)--->print 10+20=30
```

```python
class calculator:
    def add(self,no1,no2):
        print(f"{no1} + {no2} = {no1+no2}")
    def sub(self,no1,no2):
        print(f"{no1} - {no2} = {no1-no2}")
    def mul(self,no1,no2):
        print(f"{no1} * {no2} = {no1*no2}")
    def div(self,no1,no2):
        print(f"{no1} / {no2} = {no1/no2}")
```

```python
obj=calculator()
```

```python
Create a class Time. Have a method setTime which sets HH:MM:SS
Also has a method display which will display time in hours minutes and seconds.
Create a class with methods:
- increase_hour
- increase_minute
- increase_second
which will respectively increase the given unit.
```

```python
class Time:
    def set_time(self,h,m,s):
        self.h=h
        self.m=m
        self.s=s
        print(f"Time set to:{self.h}:{self.m}:{self.s}")
    def increase_hour(self):
        self.h+=1
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
    def increase_minute(self):
        self.m+=1
        if self.m>=60:
            self.h+=1
            self.m=self.m%60
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
    def increase_second(self):
        self.s+=1
        if self.s>=60:
            self.m+=1
            self.s=self.m%60
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
```

```python
t1=Time()
t2=Time()
```

```python
t1.set_time(10,59,59)
```

```python
t1.increase_second()
```

```python
t1.increase_minute()
```

```python
#__del__: Call on its own, just before the object is going to be permanently deleted, used to perform a last-minute task just before the object goes off, like closing an open file and leaving resources which you had.
```

```python
class Human:
    def __init__(self,name="unnamed",gender="unknown"):#Call on its own.
        print(f"init called:object created:{id(self)}")
        self.gender=gender
        self.name=name
    def intro(self):
        print(f"hi i am {self.name} and i am a {self.gender}")
    def __del__(self):
        print(f"{self.name} R.I.P")
```

```python
h1=Human("hritik","male")
h1.intro()
```

```python
h1=Human("kangana","female")
```

```python
#Class member only exists with the class. Hence call only with class.member. Although objects can directly access, advise not to do so.
class Human:
    population=0#This is a member of a class and can be accessed by other objects but there exists only a single copy available with the class itself.
    def __init__(self,name="unnamed",gender="unknown"):#Call on its own.
        self.gender=gender
        self.name=name
        Human.population+=1
        print(f"New human {self.name} is created Total population is {Human.population}")
    def intro(self):
        print(f"hi i am {self.name} and i am a {self.gender}")
    def __del__(self):
        Human.population-=1
        print(f"R.I.P {self.name} is deleted: Total population is {Human.population}")
```

```python
h1=Human("amar","male")
h2=Human("samar","male")
h3=Human("tamar","male")
h4=Human("uamar","male")
```
