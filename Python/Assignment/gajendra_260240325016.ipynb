```python
import getpass

def menu():
    print("----------------------")
    print("---- Secret Diary ----")
    print("----------------------")
    print("1. Read Diary")
    print("2. Write Diary")
    print("3. Exit")
    print("----------------------")


def encrypt(text, key):
    result = ""
    for ch in text:
        result = result + chr(ord(ch) + key)
    return result


def decrypt(text, key):
    result = ""
    for ch in text:
        result = result + chr(ord(ch) - key)
    return result


while True:
    menu()
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter username: ")
        key = int(getpass.getpass("Enter secret key: "))

        try:
            file = open(name + ".txt", "r")
            data = file.read()
            file.close()

            decrypted = decrypt(data, key)
            print("Your diary entry:")
            print(decrypted)

        except:
            print("Diary not found.")

    elif choice == "2":
        name = input("Enter username: ")
        key = int(getpass.getpass("Enter secret key: "))
        text = input("Write your diary entry: ")

        encrypted = encrypt(text, key)

        file = open(name + ".txt", "w")
        file.write(encrypted)
        file.close()

        print("Diary saved successfully.")
   

    elif choice == "3":
        print("Exiting program.")
        break

    else:
        print("Invalid choice")
```
